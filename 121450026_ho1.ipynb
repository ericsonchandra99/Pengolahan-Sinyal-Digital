# jawab

1. **Membuat Gelombang Sinus dan Cosinus**:
   - Buat array `t` yang berkisar dari 0 hingga 2 dengan langkah sebesar 0.0001.
   - Buat sinyal-sinyal berikut menggunakan array waktu ini:
     1. $y1 = 2 \cdot \sin(2 \pi \cdot 3 \cdot t + 0)$
     2. $y2 = 1 \cdot \cos(2 \pi \cdot 4 \cdot t + \pi/4)$
     3. $y3 = -1 \cdot \sin(2 \pi \cdot 5 \cdot t + \pi/2)$
     4. $y4 = 0.5 \cdot \cos(2 \pi \cdot 6 \cdot t + \pi)$

````python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# 1. Membuat array waktu dari 0 sampai 2 detik dengan interval 0.0001
t = np.arange(0, 2, 0.0001)

# 2. Membuat sinyal-sinyal sesuai parameter
# y1 = 2*sin(2π*3*t) - Sinus dengan amplitudo 2 dan frekuensi 3Hz
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)

# y2 = 1*cos(2π*4*t + π/4) - Cosinus dengan amplitudo 1, frekuensi 4Hz, dan phase shift π/4
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi / 4)

# y3 = -1*sin(2π*5*t + π/2) - Sinus terbalik dengan amplitudo 1, frekuensi 5Hz, dan phase shift π/2
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi / 2)

# y4 = 0.5*cos(2π*6*t + π) - Cosinus dengan amplitudo 0.5, frekuensi 6Hz, dan phase shift π
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)

````

````python
# Plot y1
plt.subplot(4, 1, 1)
plt.plot(t, y1)
plt.title('y1 = 2 sin(2π3t)')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.xlim(0, 2)

# Plot y2
plt.subplot(4, 1, 2)
plt.plot(t, y2)
plt.title('y2 = cos(2π4t + π/4)')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.xlim(0, 2)

# Plot y3
plt.subplot(4, 1, 3)
plt.plot(t, y3)
plt.title('y3 = -sin(2π5t + π/2)')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.xlim(0, 2)

# Plot y4
plt.subplot(4, 1, 4)
plt.plot(t, y4)
plt.title('y4 = 0.5 cos(2π6t + π)')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.xlim(0, 2)

plt.tight_layout()
plt.show()
````

````python
## Membuat plot
# 3. Membuat plot untuk keempat sinyal
plt.figure(figsize=(15, 8))
plt.plot(t, y1, label='Sinus dengan amplitudo 2 dan frekuensi 3Hz')
plt.plot(t, y2, label='Cosinus dengan amplitudo 1')
plt.plot(t, y3, label='Sinus terbalik dengan amplitudo 1, frekuensi 5Hz, dan phase shift π/2')
plt.plot(t, y4, label='Cosinus dengan amplitudo 0.5, frekuensi 6Hz, dan phase shift π')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sinusoidal and Cosinusoidal Wave')
plt.legend()
plt.show()
````

## 2. **Perbandingan Subplot**:
   - Buat gambar dengan 4 subplot (grid 2x2) untuk membandingkan semua sinyal secara berdampingan.
   - Setiap subplot harus berisi salah satu sinyal \( y1 \), \( y2 \), \( y3 \), dan \( y4 \).

````python
# 3. Membuat grid 2x2 untuk subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # Grid 2x2

# 4. Menambahkan sinyal pada masing-masing subplot
axs[0, 0].plot(t, y1)
axs[0, 0].set_title("Sinyal y1: 2*sin(2π*3*t+0)")

axs[0, 1].plot(t, y2)
axs[0, 1].set_title("Sinyal y2: 1*cos(2π*4*t + π/4)")

axs[1, 0].plot(t, y3)
axs[1, 0].set_title("Sinyal y3: -1*sin(2π*5*t + π/2)")

axs[1, 1].plot(t, y4)
axs[1, 1].set_title("Sinyal y4: 0.5*cos(2π*6*t + π)")

# 5. Penataan agar subplot tidak saling tumpang tindih
plt.tight_layout()

# 6. Menampilkan gambar
plt.show()
````

## 3. **Pertanyaan Analisis**:
**A. Berapa amplitudo dan frekuensi masing-masing sinyal?:** 
 - **Sinyal y1 (2*sin(2π*3*t+0))**:
     - Amplitudo: 2
     - Frekuensi: 3 Hz
     - Penjelasan: Amplitudo sinyal ini adalah 2, yang berarti puncak gelombangnya mencapai ±2. Frekuensi sinyal ini adalah 3 Hz, yang berarti ada 3 siklus per detik.
   
     - **Sinyal y2 (1*cos(2π*4*t + π/4))**:
     - Amplitudo: 1
     - Frekuensi: 4 Hz
     - Penjelasan: Amplitudo sinyal ini adalah 1, yang berarti puncak gelombangnya mencapai ±1. Frekuensi sinyal ini adalah 4 Hz, yang berarti ada 4 siklus per detik.
   
   - **Sinyal y3 (-1*sin(2π*5*t + π/2))**:
     - Amplitudo: 1
     - Frekuensi: 5 Hz
     - Penjelasan: Amplitudo sinyal ini adalah 1, dengan nilai negatif menunjukkan bahwa gelombangnya dimulai dengan posisi terbalik (dibalik vertikal). Frekuensi sinyal ini adalah 5 Hz, yang berarti ada 5 siklus per detik.
   
   - **Sinyal y4 (0.5*cos(2π*6*t + π))**:
     - Amplitudo: 0.5
     - Frekuensi: 6 Hz
     - Penjelasan: Amplitudo sinyal ini adalah 0.5, yang berarti puncak gelombangnya hanya mencapai ±0.5. Frekuensi sinyal ini adalah 6 Hz, yang berarti ada 6 siklus per detik.

   
**B. Bagaimana pergeseran fase mempengaruhi posisi gelombang? (Anda dapat mengubah nilai fase pada sinyal-sinyal yang telah dibuat sesuka anda)**
 - Pergeseran fase (phase shift) mempengaruhi posisi awal gelombang sinyal.Contohnya:
     - Pada **y2** (cos(2π*4*t + π/4)), pergeseran fase sebesar π/4 menggeser gelombang ke kanan.
     - Pada **y3** (-sin(2π*5*t + π/2)), pergeseran fase sebesar π/2 menggeser gelombang tersebut lebih jauh, dan mengubah posisi puncak gelombang.
     - **y4** (cos(2π*6*t + π)) memiliki pergeseran fase sebesar π, yang membalikkan posisi gelombang secara vertikal, yang membuatnya lebih mirip dengan gelombang sinus tetapi dimulai dengan posisi yang terbalik.

 Semakin besar pergeseran fase, semakin banyak pergeseran posisi awal gelombang, baik secara horizontal maupun vertikal, tergantung pada jenis gelombangnya (sinus atau cosinus).





````python
# C 
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# Membuat array waktu dari 0 sampai 2 detik dengan interval 0.0001
t = np.arange(0, 2, 0.0001)

# Membuat sinyal-sinyal sesuai parameter
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)        # Sinyal dengan amplitudo 2
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)  # Sinyal dengan amplitudo 1
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2) # Sinyal dengan amplitudo 1
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)  # Sinyal dengan amplitudo 0.5

# Membuat grafik untuk membandingkan amplitudo sinyal
plt.figure(figsize=(10, 6))

# Plot masing-masing sinyal dengan amplitudo berbeda
plt.plot(t, y1, label="y1: 2*sin(2π*3*t+0)", linestyle='-', color='b')
plt.plot(t, y2, label="y2: 1*cos(2π*4*t + π/4)", linestyle='--', color='r')
plt.plot(t, y3, label="y3: -1*sin(2π*5*t + π/2)", linestyle='-.', color='g')
plt.plot(t, y4, label="y4: 0.5*cos(2π*6*t + π)", linestyle=':', color='m')

# Menambahkan judul dan label
plt.title('Perbandingan Sinyal dengan Amplitudo yang Berbeda')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.grid(True)
plt.show()
````

**C. Bandingkan sinyal-sinyal dengan amplitudo yang berbeda dan diskusikan bagaimana amplitudo mempengaruhi tampilan gelombang.**
   - **Sinyal y1** dengan amplitudo 2 memiliki gelombang yang lebih tinggi dibandingkan **y2** (amplitudo 1) dan **y4** (amplitudo 0.5). Amplitudo yang lebih besar berarti puncak dan lembah gelombang akan lebih jauh dari garis nol, yang menghasilkan gelombang yang lebih 'terangkat' secara visual.
   - **Sinyal y2** dan **y3**, yang keduanya memiliki amplitudo 1, lebih kecil dibandingkan **y1**, tetapi lebih besar daripada **y4** yang memiliki amplitudo 0.5.
   - **Sinyal y4**, dengan amplitudo 0.5, akan memiliki gelombang yang lebih kecil dan lebih rapat karena perbedaan amplitudo ini.

   Jadi, amplitudo mempengaruhi 'ketinggian' gelombang: semakin besar amplitudo, semakin tinggi puncak dan lembah gelombangnya.

````python
# D 
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# Membuat array waktu dari 0 sampai 2 detik dengan interval 0.0001
t = np.arange(0, 2, 0.0001)

# Membuat sinyal-sinyal sesuai parameter dengan pergeseran fase yang berbeda
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)          # Tidak ada pergeseran fase
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)    # Pergeseran fase π/4
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)   # Pergeseran fase π/2
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)    # Pergeseran fase π

# Membuat grafik untuk membandingkan pergeseran fase
plt.figure(figsize=(10, 6))

# Plot masing-masing sinyal dengan pergeseran fase berbeda
plt.plot(t, y1, label="y1: 2*sin(2π*3*t)", linestyle='-', color='b')
plt.plot(t, y2, label="y2: 1*cos(2π*4*t + π/4)", linestyle='--', color='r')
plt.plot(t, y3, label="y3: -1*sin(2π*5*t + π/2)", linestyle='-.', color='g')
plt.plot(t, y4, label="y4: 0.5*cos(2π*6*t + π)", linestyle=':', color='m')

# Menambahkan judul dan label
plt.title('Perbandingan Sinyal dengan Pergeseran Fase yang Berbeda')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.grid(True)
plt.show()
````

**D. Bandingkan sinyal-sinyal dengan pergeseran fase yang berbeda dan diskusikan bagaimana pergeseran fase mempengaruhi tampilan gelombang.**
  - **Sinyal y2** dengan pergeseran fase π/4 akan dimulai sedikit lebih lambat dibandingkan **y1** yang tidak memiliki pergeseran fase.
   - **Sinyal y3** dengan pergeseran fase π/2 akan lebih terbelakang, dimulai dengan posisi yang terbalik dibandingkan dengan sinyal lainnya.
   - **Sinyal y4** dengan pergeseran fase π akan membalikkan arah gelombangnya, membuatnya terlihat seperti gambar cermin dari sinyal lainnya.

   Jadi, pergeseran fase mengubah posisi awal gelombang sinyal, menggesernya ke kiri atau kanan, atau bahkan membalikkan posisinya secara vertikal, tergantung pada nilai fase yang diterapkan.

## 4. **Tugas Lanjutan**:
   - Buat sinyal baru $y5$ yang merupakan kombinasi dari $y1$ dan $y2$ yaitu, $y5 = y1 + y2$.
   - Plot $y5$ dan diskusikan bagaimana kombinasi dua gelombang sinus/cosinus mempengaruhi bentuk gelombang yang dihasilkan.

````python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt


# Membuat array waktu dari 0 sampai 2 detik dengan interval 0.0001
t = np.arange(0, 2, 0.0001)

# Membuat sinyal-sinyal sesuai parameter
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)          # Sinyal 1
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)    # Sinyal 2

# Membuat sinyal y5 sebagai kombinasi dari y1 dan y2
y5 = y1 + y2

# Membuat grafik untuk memplot y5
plt.figure(figsize=(10, 6))

# Plot y5
plt.plot(t, y5, label="y5 = y1 + y2", color='b')

# Menambahkan judul dan label
plt.title('Plot Sinyal y5 = y1 + y2')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.grid(True)
plt.show()
````

### Diskusi:
- **Kombinasi dua gelombang sinus dan cosinus**: 
  - Saat dua gelombang sinusoidal dengan frekuensi dan amplitudo yang berbeda digabungkan, hasilnya adalah gelombang yang lebih kompleks.Bagian real:cos,dan Bagian imajiner: sin
  - Pada kasus ini, \( y1 \) adalah gelombang sinus dan \( y2 \) adalah gelombang cosinus dengan pergeseran fase.
  - Hasil penjumlahan \( y5 = y1 + y2 \) menghasilkan gelombang yang lebih 'terdistraksi' karena adanya perbedaan fase dan frekuensi antara kedua gelombang. Gelombang ini tidak akan memiliki bentuk sinus atau cosinus yang sempurna, melainkan akan membentuk pola baru tergantung pada seberapa besar perbedaan antara kedua gelombang tersebut.
  - Gelombang yang dihasilkan dari penjumlahan ini akan menunjukkan interaksi antara dua gelombang tersebut, yang dapat berupa penguatan atau pengurangan amplitudo pada titik-titik tertentu sesuai dengan fase relatif antara keduanya.

````python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Membuat array waktu dari 0 sampai 2 detik dengan interval 0.0001
t = np.arange(0, 2, 0.0001)

# Membuat sinyal-sinyal sesuai parameter
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)          # Sinyal 1
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)    # Sinyal 2

# Membuat sinyal y5 sebagai kombinasi dari y1 dan y2
y5 = y1 + 1j * y2  # y5 adalah sinyal kompleks, dengan y1 sebagai bagian real dan y2 sebagai bagian imaginer

# Ekstraksi bagian real dan imaginer dari y5
real_part = np.real(y5)
imag_part = np.imag(y5)

# Membuat grafik 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot bagian real dan imaginer dari y5
ax.plot(t, real_part, zs=0, zdir='y', label='Bagian Real', color='b')
ax.plot(t, imag_part, zs=1, zdir='y', label='Bagian Imaginer', color='r')

# Menambahkan judul dan label
ax.set_title('Visualisasi 3D dari Bagian Real dan Imaginer Sinyal Kompleks y5')
ax.set_xlabel('Waktu (detik)')
ax.set_ylabel('Komponen')
ax.set_zlabel('Amplitudo')

# Menambahkan legenda
ax.legend()

# Menampilkan grafik
plt.show()
````
